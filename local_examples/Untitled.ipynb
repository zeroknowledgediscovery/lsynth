```python
import pandas as pd
from quasinet.qnet import Qnet, save_qnet, load_qnet
```

```python
df=pd.read_csv('./datasets/variable72.csv',keep_default_na=False, index_col=0)
df=df.drop(['YRENTER','HHIDPN'],axis=1)
feature_names=df.columns.values
X=df.values.astype('str')
```

```python
model = Qnet(n_jobs=10,feature_names=feature_names,alpha=.05)
model.fit(X)
save_qnet(model,'modelHRS72p05',gz=True)
```

```python
df.to_csv('HRSvar72.csv')
```
