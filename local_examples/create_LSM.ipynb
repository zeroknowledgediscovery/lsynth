```python
import pandas as pd
from quasinet.qnet import Qnet, save_qnet, load_qnet
```

```python
df=pd.read_csv('./datasets/variable72.csv',keep_default_na=False, index_col=0)
df=df.drop(['YRENTER','HHIDPN'],axis=1)
df_training=df.head(5000)
feature_names=df_training.columns.values
X=df_training.values.astype('str')
```

```python
model = Qnet(n_jobs=10,feature_names=feature_names,alpha=.05)
model.fit(X)
save_qnet(model,'modelHRS72p05_train',gz=True)
```

```python
df.tail(df.index.size-5000).to_csv('HRSvar72_test.csv')
```
